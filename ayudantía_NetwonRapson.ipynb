```python
def f(x):
    return x**2 - 2

def df(x):
    return 2*x

def newton_raphson(x0, tol=1e-10, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = x - f(x)/df(x)
        if abs(x_new - x) < tol:
            print(f"Convergió en {i+1} iteraciones.")
            return x_new
        x = x_new
    raise ValueError("No convergió")

raiz = newton_raphson(1.0)
print(f"La raíz es aproximadamente: {raiz}")
# Este código implementa el método de Newton-Raphson para encontrar la raíz de la función f(x) = x^2 - 2.
# Se define la función f(x) y su derivada df(x).
# Luego, se utiliza el método de Newton-Raphson comenzando desde un valor inicial x0.
# El proceso se repite hasta que la diferencia entre dos iteraciones consecutivas sea menor que la tolerancia especificada.
# Si no se encuentra la raíz en el número máximo de iteraciones, se lanza una excepción.
# El resultado es la raíz aproximada de la función.
# El código es un ejemplo clásico de cómo implementar el método de Newton-Raphson en Python.
# Se utiliza una tolerancia de 1e-10 y un número máximo de iteraciones de 100.
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.linspace(0, 2, 400)
y_vals = f(x_vals)

x0 = 1.0
x_hist = [x0]
for _ in range(5):
    x1 = x0 - f(x0)/df(x0)
    x_hist.append(x1)
    x0 = x1

plt.plot(x_vals, y_vals, label='f(x)')
plt.axhline(0, color='black', linewidth=0.5)
for i in range(len(x_hist)-1):
    xi = x_hist[i]
    xj = x_hist[i+1]
    plt.plot([xi, xi], [0, f(xi)], 'k--')
    plt.plot([xi, xj], [f(xi), 0], 'k--')
    plt.plot(xi, f(xi), 'ro')

plt.title("Iteraciones de Newton-Raphson")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()

# Este código utiliza la biblioteca NumPy para crear un rango de valores x y calcular los valores correspondientes de f(x).
# Luego, se aplica el método de Newton-Raphson para encontrar la raíz de la función f(x) = x^2 - 2.
# Se guarda el historial de las iteraciones en la lista x_hist.
# Finalmente, se utiliza Matplotlib para graficar la función f(x) y las iteraciones del método de Newton-Raphson.
# Se dibujan líneas punteadas para mostrar las proyecciones de las iteraciones sobre el eje x y la función.
# El gráfico resultante muestra cómo el método de Newton-Raphson converge a la raíz de la función.
# El gráfico incluye etiquetas para los ejes, una leyenda y una cuadrícula para facilitar la lectura.
# El resultado es una visualización clara del proceso de convergencia del método de Newton-Raphson.
# El código es un ejemplo clásico de cómo implementar el método de Newton-Raphson en Python y visualizar su convergencia.
```
